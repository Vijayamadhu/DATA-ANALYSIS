# Complete programme for cretaing data set from TMDB movie database which is of 500 json pages when we view it from jason online viewer
# Using for loop to iterate through all the pages and append all the data to the corresponding empty list


```python
import requests
import pandas as pd
```

```python
movie_dataset=pd.DataFrame()
for j in range(1,500):
    response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=a7e784951966bf44afb76db195d4c9f6&language=en-US&page={}'.format(j)).json()
    title=[]
    vote_average=[]
    popularity=[]
    vote_count=[]
    original_language=[]
    overview=[]
    release_date=[]
    for i in response['results']:
        try:
            release_date.append(i['release_date'])
        except KeyError:
            continue 

        title.append(i['title'])
        vote_average.append(i['vote_average'])
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])
        
    
    d={'title':title,'vote_average':vote_average,'popularity':popularity,'vote_count':vote_count,'original_language':original_language,'overview':overview,'release_date':release_date}
    d = dict([ (k,pd.Series(v)) for k,v in d.items() ])
    df=pd.DataFrame(d) 
    movie_dataset=pd.concat([df,movie_dataset],ignore_index=True)
 
```
